```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import statsmodels.api as sm
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("T:\\PYTHON-programs\\machine-learning\\linear_regression\\first_linear_regression\\real_estate_price_size.csv")
print(data)
```

```python
sns.pairplot(data)
```

### Univariate Linear Regression

##### OLS (Ordinary Least Squares) Linear Regrssion using StatsModel

```python
y = data['price']
x1 = data['size']

x = sm.add_constant(x1)
reg_linear = sm.OLS(y, x)
results = reg_linear.fit()

print(results.summary())
```

##### Linear Regression using sklearn 

```python
x = data["size"]
y = data["price"]

x = x.values.reshape(100,1)

reg_model = LinearRegression()
reg_model.fit(x,y)
```

```python
print(reg_model.score(x,y))
print(reg_model.coef_)
print(reg_model.intercept_)
```

##### Plotting the regression line 

```python
plt.scatter(x,y,marker = "h")
plt.xlabel("Area (Sq.ft)")
plt.ylabel("Price")
yhat = (reg_model.coef_ * x) + reg_model.intercept_
plt.plot(x,yhat, lw = 2, c = "red")
plt.show()
```
